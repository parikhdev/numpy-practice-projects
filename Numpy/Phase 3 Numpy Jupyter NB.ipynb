```python
import numpy as np
import matplotlib.pyplot as plt

# Advanced Topics with Practical Questions
```

```python
# Data Structure -> [Restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Briyani
    [2, 120000, 160000, 190000, 240000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000]  # Chai Point
])

# Shape[size of the data]
print(f"Sales data shape is: {sales_data.shape}")
print(f"Sample data for first 3 restaurant are:\n{sales_data[:3]}")
```

```python
# Yearly total -> np.sum(data[shape], axis =0/1)
Yr_total = np.sum(sales_data[:, 1:], axis = 0)
Rest_total = np.sum(sales_data[:, 1:], axis = 1)
print(f"The yearly total of all years are:\n{Yr_total}\nThe Restaurant total of all restaurants are:\n{Rest_total}")

# Min. Sales -> np.min(data[shape], axis =0/1)
Min_Sales = np.min(sales_data[:, 1:], axis = 0)
print(f"Min. Sales are:\n{Min_Sales}")

# Max. Sales -> np.max(data[shape], axis =0/1)
Max_Sales = np.max(sales_data[:, 1:], axis = 0)
print(f"Max. Sales are:\n{Max_Sales}")

# Average Sales -> np.mean(data[shape], axis =0/1)
Avg_Sales = np.mean(sales_data[:, 1:], axis = 0)
print(f"Avg. Sales are:\n{Avg_Sales}")

# Cumulative Sales -> np.cumsum(data[shape], axis =0/1)
Cum_Sales = np.cumsum(sales_data[:, 1:], axis = 1)
print(f"Cumulative Sales are:\n{Cum_Sales}")
```

```python
# Plot Graph -> plt.figure(figsize =(R,C))
plt.figure(figsize =(4,5))
plt.plot(np.mean(Cum_Sales, axis=0))
plt.title("Average Cumulative Sales across all Restaurants: ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Vector Addition -> vector1 + vector2
v1 = np.array([1,2,34,50,8])
v2 = np.array([31,21,1,9,4])
print(f"The Vector Addition is:\n{v1+v2}")

# Vector Multiplication -> vector1 * vector2
print(f"The Vector Multiplication is:\n{v1*v2}")

# Dot Product -> np.dot((vector1, vector2))
v3 = np.dot(v1,v2)
print(f"The Vector Dot Product is:  {v3}")

# Angle -> np.arccos(np.dot(vector1, vector2)/(np.linalg.norm(vector1) * np.linalg.norm(vector2)))

Angle = np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1) * np.linalg.norm(v2)))
print(f"The Angle is:\n{Angle}")
```
